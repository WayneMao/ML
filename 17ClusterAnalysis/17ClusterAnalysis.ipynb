## 17 基于聚类分析的个性化影响
### 17.1 理解数字化营销场景

**营销管理中的用户分群。**

只要接触过营销业务，立刻就能想到什么是用户分群了。 设想一下，公司CRM系统里有大量的用户资源。为了迎接特殊的节日，公司打算搞一次用户回馈活动，该如何设计呢？

一种简单的方式是针对于每一个人采用同样的营销手段，比如优惠一定的金额。但这样的方法显然是不具备个性化的，而且我们要知道不同的客户想要的东西是不一样的。所以，更理想的方式是：根据用户的特性给用户做分层，然后做针对性的营销。比如对于A类用户重点推荐高端产品，对于B类用户重点推荐中低端产品等等。

那对于用户如何分层呢？ 这是摆在每一位营销人员面前的问题！不同的方法会带来不一样的效果。  
对于用户分层上有几种常见的方法。其中最经典也最简单的方法叫作RFM模型，是通过3个不同维度给用户做分层。  
比RFM稍微复杂一点的无非就是K-Means聚类算法了。

### 17.2 RFM模型与用户分层
**RFM模型**一个在营销场景里非常著名的方法论。


这里的RFM分别代表R(Recency)，F(Frequency)和M(Moneytary)，我们逐一来看一下。


**Recency**: 代表的是用户最近一次购买时间，比如用户最后一次购买是10天前，或者20天前。从Recency可以部分看出用户价值的，比如一天前光顾饭店的客户相比一个月前光顾的客户价值更大一些。 所以这个值越小，用户价值就越高。

**Frequency**：代表的是用户在过去一段时间购买次数。比如一个用户在过去一年在店里购买了10次，则Frequency值为10。如果一个用户的购买次数越多，那他的用户价值会越高。

**Moneytary**：代表的是用户在过去一段时间内所花费的总金额。比如一个用户过去一年在某个店里花费了10万，这就是Moneytary值。这个值越大说明用户价值就越高。

所以，当我们使用RFM模型的时候第一步是先把每一位用户的R,F,M的值算出来。这个过程通过统计计算就可以获得，按照用户的维度，把每一个用户过去一段时间内所有的交易记录整理一下就可以了。

计算完之后就可以得到每一位用户的R,F,M的值。之后根据这三个值就可以对用户分群了。

---
根据R,F,M值分群的经典方法：

一种方法是我们先把每一值映射到了不同的bin, 然后根据自定义的规则对人群做了分群。  
使用quantile就可以了。 把一组数据通过quantile分成25%， 50%， 75%点， 这样就可以分成4块了，然后每一块分别赋值，1，2，3，4就可以了。 取quantitle的时候可以直接使用 ```pd.qcut```函数。  
这种方法的好处在于简单，而且可解释性比较强。但缺点是引入了很多人工的处理和规则。

另外一种方法是使用K-means等聚类算法，直接应用在R,F,M数据或者原始数据中。使用K-means算法时需要注意的一点是：特征的归一化，因为每个特征维度的大小会有很大区别。

另外，使用K-Means算法的时候需要合理地选择K值，这部分在上一章里已经涉及过，在项目里也提供了详细的描述。

### 17.3 聚类算法的评估方法
在本节，我们简单来学习一下如何评估这种方法。

首先，聚类分析是无监督算法，所以做聚类分析之前是没有标签数据的。在没有标签数据的情况下，只能依靠从数据本身的理解或者靠经验来设定分群。但随着活动的正式进行，我们可以逐渐获得一些反馈数据的。比如A类客户的转化率有多少，B类客户的转化率有多少等等。根据用户的反馈数据，我们就可以调整分群策略了。但这个过程也需要一些技巧。

需要思考的一点是：如何使用最小成本来尝试并调整分群策略？   
一般采用的很重要的技术是A/B测试。比如我们手里有两种分群方案A和B。A方案是目前来讲比较有效的而且稳定的，B方案是想要尝试的而且很可能比较有效的。那这时候我们可以通过快速验证的方式来评估B方案相比A方案的优势。比如对于采样的小部分用户我们实施B方案，之后再看一下效果。

其实任何的AI产品都离不开**A/B测试**，核心是如何用最小成本去验证某个方案的优越性。

最后的问题是，如何去评估呢？ 这个其实要看业务场景了！ 比如营销里我们关注的是转化率，那指标就定转化率。
如果关注的新增用户，那这就是我们的结果指标。所以具体的指标，需要看当前业务，没有一个标准的答案。
